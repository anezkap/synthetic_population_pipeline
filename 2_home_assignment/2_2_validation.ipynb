```python
import pandas as pd
import geopandas as gpd
```

```python
# Load data
df = pd.read_csv("output/employed_population_with_home_address.csv")
```

```python
# Check for missing home_x and home_y values
missing_coords = df[df["home_x"].isnull() | df["home_y"].isnull()]
print(f"Num records with missing coordinates: {len(missing_coords)}")

# Count the number of people per address
people_per_address = df.groupby(["home_x", "home_y"]).size().describe()

# Order by the number of people per address in descending order
people_per_address = people_per_address.sort_values(ascending=False)

# Print the ordered results
print(f"Number of people per address (sorted):\n{people_per_address}")
```

```python
# Get the address with the most people
most_populated_address = df.groupby(["home_x", "home_y", "assigned_sector"]).size()
```

```python
print(most_populated_address.sort_values(ascending=False).head(10))
```

```python
# Used to check the ZZZZ sectors in the synthetic population
csv_file = "../1_synthetic_population/output/employed_population.csv"
df = pd.read_csv(csv_file)

sectors = df['sector'].unique()
print(f"Number of unique sectors: {len(sectors)}")

# Filter to only get the ZZZZ sectors
zzzz_sectors = [s for s in sectors if s.endswith("ZZZZ")]
print(f"Number of ZZZZ sectors: {len(zzzz_sectors)}")

print(f"ZZZZ sectors: {zzzz_sectors}")
```
