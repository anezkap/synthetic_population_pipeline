# Remove long distance participants who use public transport

These have the travel mode fixed anyway and are teleported, so they don't need to be simulated in MATSim

```python
import pandas as pd
```

```python
workers = pd.read_csv('../10_subpopulation_tags/output/all_active_workers_final.csv')
```

```python
workers_reduced = workers[ ~( (workers['subpopulation'] == 'long_distance') & (workers['assigned_mode'] == 'pt') ) ]
workers_ld_pt = workers[ (workers['subpopulation'] == 'long_distance') & (workers['assigned_mode'] == 'pt') ]

print(f"Number of workers before removal: {len(workers)}")
print(f"Number of workers after removal: {len(workers_reduced)}")
print(f"Number of workers removed: {len(workers) - len(workers_reduced)}")
print(f"Percentage of workers removed: {(len(workers) - len(workers_reduced)) / len(workers) * 100:.2f}%")
print(f"Number of long-distance PT workers Brussels residents: {len(workers_ld_pt[workers_ld_pt['lives_in_brussels']])}")
print(f"Percentage of long-distance PT workers who are Brussels residents: {len(workers_ld_pt[workers_ld_pt['lives_in_brussels']]) / len(workers[workers['lives_in_brussels']]) * 100:.2f}%")
print(f"Number of long-distance PT workers who are Brussels workers: {len(workers_ld_pt[workers_ld_pt['works_in_brussels']])}")
print(f"Percentage of long-distance PT workers who are Brussels workers: {len(workers_ld_pt[workers_ld_pt['works_in_brussels']]) / len(workers[workers['works_in_brussels']]) * 100:.2f}%")
```

```python
print("All workers:")
mode_split = (
    workers["assigned_mode"]
    .value_counts(dropna=False)
    .rename_axis("mode")
    .to_frame("count")
)

mode_split["share_pct"] = (mode_split["count"] / mode_split["count"].sum() * 100).round(2)
print(mode_split)

print("\nReduced workers:")
mode_split = (
    workers_reduced["assigned_mode"]
    .value_counts(dropna=False)
    .rename_axis("mode")
    .to_frame("count")
)

mode_split["share_pct"] = (mode_split["count"] / mode_split["count"].sum() * 100).round(2)
print(mode_split)
```

```python
workers_reduced.to_csv('output/reduced_workers.csv', index=False)
workers_ld_pt.to_csv('output/long_distance_pt_workers.csv', index=False)
```
