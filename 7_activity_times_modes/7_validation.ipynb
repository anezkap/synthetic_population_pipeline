```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pop = pd.read_csv('output/workers_with_activity.csv')
```

```python
print("Modal split for work trips:")
print(pop['assigned_mode'].value_counts()/len(pop) * 100)
```

```python
print("\nCar ownership:")
print(pop['has_car'].value_counts()/len(pop) * 100)

print("\nCompany car ownership:")
print(pop['has_company_car'].value_counts()/len(pop) * 100)
```

```python
counts = pop.groupby(['dist_bracket', 'assigned_mode']).size().unstack(fill_value=0)

counts_pct = counts.div(counts.sum(axis=1), axis=0) * 100
ax_pct = counts_pct.plot(kind='bar', stacked=True, figsize=(10,6))
ax_pct.set_xlabel('dist_bracket')
ax_pct.set_ylabel('percentage (%)')
ax_pct.set_title('Modal split by distance bracket (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Map education to focused groups
ISCED11_TO_FOCUS32 = {
    'ISCED11_1': 'lower',
    'ISCED11_2': 'lower',
    'ISCED11_3': 'lower',
    'ISCED11_4': 'lower',
    'ISCED11_5': 'higher',
    'ISCED11_6': 'higher',
    'ISCED11_7': 'higher',
    'ISCED11_8': 'higher',
    'UNK': 'unknown',
    'NONE': 'lower',
}

# Create a new column with grouped education
pop['education_grouped'] = pop['education'].map(ISCED11_TO_FOCUS32)

# Calculate modal split by education
counts_edu = pop.groupby(['education_grouped', 'assigned_mode']).size().unstack(fill_value=0)
counts_edu_pct = counts_edu.div(counts_edu.sum(axis=1), axis=0) * 100

# Plot
ax_edu_pct = counts_edu_pct.plot(kind='bar', stacked=True, figsize=(10, 6))
ax_edu_pct.set_xlabel('education')
ax_edu_pct.set_ylabel('percentage (%)')
ax_edu_pct.set_title('Modal split by education (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
departure_hours = pop['departure_home_to_work'] / 60

plt.figure(figsize=(10, 6))
plt.hist(departure_hours, bins=30, edgecolor='black')
plt.xlabel('Departure from home to work (hours after midnight)')
plt.ylabel('Count')
plt.title('Histogram of departure_home_to_work')
plt.tight_layout()
plt.show()
```

```python
print(f"Number of unique departure times: {len(pop['departure_home_to_work'].unique())}")
print(departure_hours.value_counts().head())
```

```python
departure_hours = pop['work_duration'] / 60

plt.figure(figsize=(10, 6))
plt.hist(departure_hours, bins=15, edgecolor='black')
plt.xlabel('Work duration (hours)')
plt.ylabel('Count')
plt.title('Histogram of work_duration')
plt.tight_layout()
plt.show()
```

```python
# Plot the historgram of commute distances
# Cut off at 100 km for better visualization
pop = pop[pop['commute_dist_km'] <= 100]
plt.figure(figsize=(10, 6))
plt.hist(pop['commute_dist_km'], bins=30, edgecolor='black')
plt.xlabel('Commute distance (km)')
plt.ylabel('Count')
plt.title('Histogram of commute distances')
plt.tight_layout()
plt.show()
```
