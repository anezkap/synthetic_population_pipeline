# Filters cars per statistical sector dataset

Only keeps data for Brussels and renames columns

```python
import pandas as pd
```

```python
# Filter to get only data for Brussels

# Load car ownership data
cars_file = "input_data/cars_per_hh_ss_2021.xlsx"
df = pd.read_excel(cars_file)

# Display columns to verify
print("Available columns in cars data:")
print(df.columns.tolist())
print(f"\nTotal records: {len(df)}")

# Filter for Brussels province (CD_REFNIS = 210XX)
df_brussels = df[df["CD_REFNIS"].astype(str).str.startswith("210")].copy()
print(f"Records for Brussels: {len(df_brussels)}")

# Keep only sector, 
# Mapping: cd_sector for sector, total_huisH for # of household, total_wagens for # of cars
df_cleaned = df_brussels[["cd_sector", "total_huisH", "total_wagens"]].copy()

# Rename employment column to CD_EMPLOYMENT for consistency
df_cleaned.rename(columns={"cd_sector": "CD_SECTOR"}, inplace=True)
df_cleaned.rename(columns={"total_huisH": "NUM_HOUSEHOLDS"}, inplace=True)
df_cleaned.rename(columns={"total_wagens": "NUM_CARS"}, inplace=True)

# Save cleaned data
output_file = "input_data/cars_per_hh_ss_2021_cleaned.csv"
df_cleaned.to_csv(output_file, index=False)
print(f"\nCleaned data saved to {output_file}")
print(f"Final shape: {df_cleaned.shape}")
```

# Print quick statistics

```python
# Sum the number of households and cars
total_households = df_cleaned["NUM_HOUSEHOLDS"].sum()
total_cars = df_cleaned["NUM_CARS"].sum()

print(f"Total households: {total_households}")
print(f"Total cars: {total_cars}")
print(f"Cars per household: {total_cars / total_households:.2f}")

# Calculate sector statistics
num_sectors = df_cleaned["CD_SECTOR"].nunique()
avg_households_per_sector = df_cleaned["NUM_HOUSEHOLDS"].mean()
max_households = df_cleaned["NUM_HOUSEHOLDS"].max()
min_households = df_cleaned["NUM_HOUSEHOLDS"].min()

print(f"Number of sectors: {num_sectors}")
print(f"Average households per sector: {avg_households_per_sector:.2f}")
print(f"Max households in a sector: {max_households}")
print(f"Min households in a sector: {min_households}")
```
